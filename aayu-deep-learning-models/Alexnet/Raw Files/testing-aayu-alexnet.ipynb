```python
labels = {'Arjun': 0, 'Guava': 1, 'Jamun': 2, 'Jatropa': 3, 'Lemon': 4, 'Mango': 5}
```

```python
inputImg = './Dataset/Prediction-Dataset/1.jpg'
```

```python
# from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
img = plt.imshow(mpimg.imread(inputImg))
```

```python
# Process Image
borderType = cv2.BORDER_CONSTANT
colorBorder = [255, 255, 255]
increaseSize = 750

read = cv2.imread(inputImg)
width = int(read.shape[1])
height = int(read.shape[0])
print(str(width)+'x'+str(height))

if (height > width) :
    h = int((height-width)/2)
    hmax = int(h + increaseSize)
    output = cv2.copyMakeBorder(read, increaseSize, increaseSize, hmax, hmax, borderType, None, colorBorder)
else :
    w = int((width-height)/2)
    wmax = int(w + increaseSize)
    output = cv2.copyMakeBorder(read, wmax, wmax, increaseSize, increaseSize, borderType, None, colorBorder)
        
# Resize to Alexnet Image Size
width = int(227)
height = int(227)
imgSize = (width, height)
outputImg = cv2.resize(output, imgSize)

folderPath = "./Dataset/Prediction-Dataset/"
imgName = ('1-resize.jpg')
filename = folderPath+imgName
cv2.imwrite(filename, outputImg)
resizeImg = folderPath+imgName
```
